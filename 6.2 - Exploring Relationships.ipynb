6. Finding out how many universities are in each country

```python
#Number of universities by country
university_by_country= df.groupby('Country')['Institution'].nunique()
university_by_country
```

It is clear that the United States has the most universities in the world, but I would like to find out which country has the best scores, on average.

```python
#find the average university scores by country
university_average_by_country= df.groupby('Country')['Score'].mean()
university_average_by_country
```

Based on the average university scores for each country, the United States has the second highest score, second only to Israel, but Israel only has seven Universities and the United States has 231. 

HYPOTHESIS: The Universities that have the highest scores are located in the United States, therefore, the United States has the best Universities.
